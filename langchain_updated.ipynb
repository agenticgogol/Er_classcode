```python
# 4. Vector DB
# In Memory  (RAM) - data is lost when the session restarts ( FAISS, CromaDB)
# Local (Hard disk) - here u store teh database as a file ( FAISS, CromaDB)
# Cloud hosted - hosted by vendors - scaliing , multi team access , APIs (Pinecone, Weaviate, Milvus)
# AWS - Opensearch
# Azure - Ai Search
# GCP vertex AI vector search

# Weaviate has a local support as well as cloud support 
# Pinecone has only cloud support
# FAISS, Croma has only local support

# Every vector DB needs to have index 

# Semantic Search 
# - Exact - FlatIndex [ Every chunk that is stored will be compared and result will be provided]
# - Approximate - Grapg based and Inverted index based 
# Hierarchical Navigable Small World (HNSW) - Graph based 
# and Inverted File with IVF Flat Indexing (IVFFlat)# 

# Similarity match
# Cosine Similarityy
# Euclidean similarity ( L2) 
# Jaccard similarity 
```

```python
# Similarity Matching
embeddings1 = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")

from sklearn.metrics.pairwise import cosine_similarity
documents = ["Capital of USA is Washington", 
             "Capital of Germany is Berlin",
             "President of USA is Donald Trump", 
             "Narendra Modi is prime minister of India",
             "Narendra Modi was Gujarat's CM earlier"]


#my_query = "What is Agentic AI?"
my_query = "What is Prime minister of India?"

document_embedded = embeddings1.embed_documents(documents)
query_embedded = embeddings1.embed_query(my_query)
cosine_similarity([query_embedded],document_embedded)
# Cos theta = angle between two vector = cos 0 = 1
# if cost theta is high or closer to 1 - it is closer match

```

```python
from sklearn.metrics.pairwise import euclidean_distances
euclidean_distances([query_embedded],document_embedded)
```

```python
len(query_embedded) # This length is specific to embedding
# For a given embedding model , the length of the embedding will remain constant irrrespective of text / chunk size
# Chunk size / text size = 1000 , 10000, 200000
# irrespective of that it will be represented by a 384 length vector as long as you are using this embedding model
#If you change ur embvedding model , you will see this length this change
```

```python
embeddings1 = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
index = faiss.IndexFlatL2(384) # Exact Match(=Flat) , Euclidean Distance(=L2), [size of vector = 384)]
vector_store = FAISS(
    embedding_function = embeddings1,
    index = index,
    docstore = InMemoryDocstore(),
    index_to_docstore_id={}
)
vector_store.add_texts(documents)
results = vector_store.similarity_search("What is Prime minister of India?",k=2)
results 



```

```python
# End to End RAG Example 
# Step 1 - 
from langchain_community.document_loaders import PyPDFLoader
loader =  PyPDFLoader('human-rights-statement.pdf')
pdf_docs =  loader.load()

# Step 2
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 20
)
split_docs = splitter.split_documents(pdf_docs)
print(len(split_docs))

# Step 3
embeddings1 = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2") # Define embedding model
index = faiss.IndexFlatIP(384) # Exact Match(=Flat) , Cosine Similarity(=IP), [size of vector = 384)]
vector_store = FAISS(
    embedding_function = embeddings1,
    index = index,
    docstore = InMemoryDocstore(),
    index_to_docstore_id={}
)
vector_store.add_documents(documents = split_docs)
retriever = vector_store.as_retriever(
    search_kwargs = {"k":5}
)
# Step 4: 
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model = "o1-mini")

```

```python
# Step 4: production style writing


from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
#prompt = hub.pull("rajkstats/finance-rag-prompt-non-reasoning")
import pprint
pprint.pprint(prompt.messages)
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
rag_chain = (
    {"context":retriever | format_docs,
     "question":RunnablePassthrough()}
     |prompt
     |model
     |StrOutputParser()
)
#rag_chain.invoke(" What is the Anti-Bribery and Anti-Corruption policy?")
rag_chain.invoke(" Can you take three important keywords from the last conversation ?")
```
